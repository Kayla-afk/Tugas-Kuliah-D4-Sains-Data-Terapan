**6.** Jika sebuah barang berharga 750.000 dan mendapat diskon 45%, hitunglah harga akhirnya setelah diskon.

```python
barang = 750000
diskon = 0.45
harga_diskon = barang * (1 - diskon)
print(harga_diskon)
```
