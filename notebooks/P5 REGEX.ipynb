```python
#exercise 2
import re

file_path = "/content/berita.txt"

def extract_abbreviations(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()
            abbreviations = re.findall(r'\b[A-Z]{2,}(?:\.[A-Z]+)*\b', text)
        return abbreviations
    except FileNotFoundError:
        return "File not found. Please check the file path."
    except UnicodeDecodeError:
        return "Encoding error. Try changing the encoding to match your file's format."

abbreviations = extract_abbreviations(file_path)

if isinstance(abbreviations, list):
    print("Extracted Abbreviations:", abbreviations)
else:
    print(abbreviations)
```

```python
#exercise 3
with open('/content/berita.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

titles = [line.strip() for line in lines if line.strip()]  # Remove empty lines and spaces

titles_sorted = sorted(titles)

for title in titles_sorted:
    print(title)
```

```python
#exercise 4

with open('/content/berita.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

names = []

for line in lines:
    if "inaugurated professor" in line.lower():
        name = line.split("inaugurated professor")[0].strip()
        if name:
            names.append(name)

names_sorted = sorted(names)

for name in names_sorted:
    print(name)
```

```python
#exercise 5
import pandas as pd
import re

data = pd.read_csv('/content/reviews_dataset_updated.csv')

def clean_review(review):

    if isinstance(review, str):
        review = re.sub(r'[^\w\s]', '', review)
        review = re.sub(r'\d+', '', review)
        review = review.lower()
        return review
    return ''

data.to_csv('/content/reviews_dataset_updated.csv', index=False)
print("Data cleaned and saved to cleaned.csv successfully!")
```
