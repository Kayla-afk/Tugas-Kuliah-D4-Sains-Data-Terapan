# **CONTOH 3: BERAT BADAN**

```python
import pandas as pd

# Data survei
data = [
    ['muda', 'over', 'tidak'],
    ['muda', 'rata-rata', 'tidak'],
    ['muda', 'kurang', 'tidak'],
    ['muda', 'kurang', 'tidak'],
    ['muda', 'kurang', 'tidak'],
    ['muda', 'rata-rata', 'tidak'],
    ['muda', 'rata-rata', 'ya'],
    ['muda', 'over', 'ya'],
    ['muda', 'over', 'ya'],
    ['muda', 'rata-rata', 'ya'],
    ['tua', 'over', 'ya'],
    ['tua', 'rata-rata', 'ya'],
    ['tua', 'rata-rata', 'ya'],
    ['tua', 'kurang', 'tidak'],
    ['tua', 'kurang', 'tidak'],
    ['tua', 'kurang', 'ya'],
    ['tua', 'rata-rata', 'ya'],
    ['tua', 'over', 'ya'],
    ['tua', 'over', 'ya'],
    ['tua', 'rata-rata', 'ya']
]

# Buat DataFrame
df = pd.DataFrame(data, columns=['usia', 'berat_badan', 'hipertensi'])

# Hitung probabilitas dasar
total = len(df)
p_hipertensi = df['hipertensi'].value_counts(normalize=True).to_dict()

# Hitung probabilitas gabungan P(usia, berat_badan)
joint_counts = df.groupby(['usia', 'berat_badan']).size()
joint_probs = joint_counts / total

# Hitung probabilitas bersyarat P(hipertensi | usia, berat_badan)
def conditional_probability(usia, berat_badan, target='ya'):
    subset = df[(df['usia'] == usia) & (df['berat_badan'] == berat_badan)]
    if len(subset) == 0:
        return 0
    count_target = subset[subset['hipertensi'] == target].shape[0]
    return count_target / len(subset)

# Contoh: P(hipertensi='ya' | usia='tua', berat_badan='over')
usia_input = 'tua'
berat_input = 'over'
p_conditional = conditional_probability(usia_input, berat_input)

print(f"Probabilitas seseorang dengan usia '{usia_input}' dan berat badan '{berat_input}' mengalami hipertensi adalah: {p_conditional:.2f}")
```

# **CONTOH 3: BERMAIN TENIS**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Data cuaca dan keputusan bermain tenis
data = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong'],
    'PlayTennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

# Buat DataFrame
df = pd.DataFrame(data)

# Encode fitur kategorikal
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])

# Pisahkan fitur dan target
X = df.drop('PlayTennis', axis=1)
y = df['PlayTennis']

# Split data untuk evaluasi
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi dan latih model Naive Bayes
model = CategoricalNB()
model.fit(X_train, y_train)

# Prediksi dan evaluasi
y_pred = model.predict(X_test)
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nLaporan klasifikasi:\n", classification_report(y_test, y_pred))
```
