```python
#KAYLA NUANSA CERIA_3324600023_UTS PEMROGRAMAN SEMESTER 2
'''
Program untuk menganalisis frekuensi aktivitas user pada log aktivitas terlampir
'''
import re
file_path = "C:/Users/lenovo/Downloads/log_aktivitas.txt"

def extract_abbreviations(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()
            abbreviations = re.findall(r'\b[A-Z]{2,}(?:\.[A-Z]+)*\b', text)
        return abbreviations
    except file: 
        return "File log_aktivitas.txt berhasil dibaca"
    except FileNotFoundError:
        return "File log.txt tidak ditemukan."
   
def FrekuensiTiapAktivitas(prompt, low_limit, high_limit):
    while True:
        try:
            value = int(input(prompt))
            if low_limit <= value <= high_limit:
                return value
            else:
                print(f"Error: the value is not within permitted range ({low_limit}..{high_limit})")
        except ValueError:
            print("Error: wrong input")
            result = FrekuensiTiapAktivitas("nama file ringkasan", 1,2,3)
print(f"You entered a valid frequence: {result}")

def menu():
    print("\nMenu:")
    print("1. Masukkan nama file yang ingin dibaca")
    print("2. Daftar aktifitas yang ditemukan")
    return input("Pilih opsi (1/2): ")

def tulis_ke_file():
    with open("C:/Users/lenovo/Downloads/log_aktivitas.txt", 'w') as file:  
        teks = input("Masukkan teks: ")
        file.write(teks)
        print("Teks berhasil ditulis.")

def baca_dari_file():
    try:
        with open("C:/Users/lenovo/Downloads/log_aktivitas.txt", 'r') as file:  
            print("Isi file:")
            print(file.read())
    except FileNotFoundError:
        print("File tidak ditemukan.")

while True:
    pilihan = menu()
    if pilihan == '1':
        tulis_ke_file()
    elif pilihan == '2':
        baca_dari_file()
    elif pilihan == '3':
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak valid.")
```
