```python
#exercise 1
def input_integer_in_range(prompt, low_limit, high_limit):
    while True:
        try:
            value = int(input(prompt))
            if low_limit <= value <= high_limit:
                return value
            else:
                print(f"Error: the value is not within permitted range ({low_limit}..{high_limit})")
        except ValueError:
            print("Error: wrong input")
result = input_integer_in_range("Enter an integer value: ", 10, 50)
print(f"You entered a valid value: {result}")
```

```python
#exercise 2
def add_numbers():
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        result = num1 + num2

        print(f"The result of {num1} + {num2} is {result}")

    except ValueError:
        print("Error: Please enter valid numeric values.")

add_numbers()
```
