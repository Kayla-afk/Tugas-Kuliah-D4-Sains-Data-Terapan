```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Membaca Dataset dan Menampilkan
dataset = pd.read_csv('/content/titanic.csv')
display(dataset.head())
```

```python
#2. Menghitung baris dan kolom dataset
rows, cols = dataset.shape
print(f"Jumlah baris (rows): {rows}")
print(f"(Jumlah kolom (cols): {cols})")
```

```python
#3. Mengambil dan menampilkan dataset kolom fitur (Name, Sex, Age, Pclass, Fare)
data = dataset [['Name', 'Sex', 'Age', 'Pclass', 'Fare']].copy()
display(data.head())
```

```python
#4. Mengambil dan menampilkan kolom kelas (survived)
kelas = dataset['Survived']
display(kelas.head())
```

```python
#5. Menambahkan fitur relatives
data['Relatives'] = dataset['SibSp'] + dataset ['Parch']
display(data.head())
```

```python
#6. Menghitung berapa penumpang pada masing-masing nilai Pclass
jumlah_pclass = dataset['Pclass'].value_counts().sort_index()
print("Jumlah penumpang per Pclass:")
print(jumlah_pclass)
```

```python
#7. Menghitung berapa penumpang pada masing-masing nilai Sex
jumlah_sex = dataset['Sex'].value_counts()
print("\nJumlah penumpang per Sex:")
print(jumlah_sex)
```

```python
#8. Menghitung berapa penumpang yang tidak selamat dan selamat
survived_pclass = pd.crosstab(dataset['Pclass'], dataset['Survived'])
print("\nJumlah penumpang Survived (0 = Tidak Selamat, 1 = Selamat) per Pclass:")
print(survived_pclass)
```

```python
#9. Visualisasi dengan warna yang berbeda untuk class (sumbu x= urutan data, sumbu y= Sex)
plt.figure(figsize=(10, 3))
sns.scatterplot(x=dataset.index, y=dataset['Sex'], hue=dataset['Survived'], palette='Set1', alpha=0.6)
plt.title('Sebaran Data: Urutan Data vs Sex (Berdasarkan Survived)')
plt.xlabel('Urutan Data')
plt.ylabel('Sex')
plt.show
```

```python
#10. Visualisasi dengan warna yang berbeda untuk class (sumbu x=urutan data dan sumbu y=Age)
dataset_age_clean = dataset.dropna(subset=['Age'])

plt.figure(figsize=(12, 5))
sns.scatterplot(x=dataset_age_clean.index, y=dataset_age_clean['Age'], hue=dataset_age_clean['Survived'], palette='Set1', alpha=0.7)
plt.title('Sebaran Data: Urutan Data vs Age (Berdasarkan Survived)')
plt.xlabel('Urutan Data')
plt.ylabel('Age')
plt.show()
```
