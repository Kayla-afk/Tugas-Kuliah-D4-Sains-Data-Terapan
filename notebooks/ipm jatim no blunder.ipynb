```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import pandas as pd

# Baca dataset Excel
data = pd.read_excel("/content/IPM Jatim.xlsx")

# Pilih fitur numerik untuk clustering
X = data[["Indeks Kesehatan", "Indeks Pendidikan", "Indeks Ekonomi"]]

# Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Terapkan DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=3)
clusters = dbscan.fit_predict(X_scaled)

# Tambahkan hasil cluster ke dataset
data["Cluster"] = clusters

# Visualisasi hasil cluster
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters, cmap="viridis", s=50)
plt.xlabel("Indeks Kesehatan (scaled)")
plt.ylabel("Indeks Pendidikan (scaled)")
plt.title("Clustering DBSCAN IPM Jatim")
plt.colorbar(label="Cluster")
plt.show()

# Tampilkan hasil clustering dalam tabel
from IPython.display import display
display(data.head(20))  # tampilkan 20 baris pertama
```
