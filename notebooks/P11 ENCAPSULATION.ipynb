Lakukan enkapsulasi agar user tidak bisa merubah nilai atribut login_attempt selain menggunakan method increment_login_attempt() dan reset_login_attempt().

```python
#LATIHAN 7
class Restaurant:
    def __init__(self, name, cuisine_type):
        self.name = name
        self.cuisine_type = cuisine_type
        self.__menu = {}

    def describe_restaurant(self):
        print(f"Restoran {self.name} berada di {self.cuisine_type}.")

    def open_restaurant(self):
        print(f"Restoran {self.name} sudah dibuka!")

    def display_menu(self):
        print(f"Daftar menu untuk restoran {self.name}:")
        if self.__menu:
            for dish, price in self.__menu.items():
                print(f" - {dish}: Rp{price:,}")
        else:
            print(" Menu belum tersedia.")

    def add_menu(self, dish, price):
        self.__menu[dish] = price
        print(f"Menu '{dish}' berhasil ditambahkan dengan harga Rp{price:,}.")

    def remove_menu(self, dish):
        if dish in self.__menu:
            del self.__menu[dish]
            print(f"Menu '{dish}' berhasil dihapus.")
        else:
            print(f"Menu '{dish}' tidak ditemukan di restoran {self.name}.")


class User:
    __max_login_attempt = 3

    def __init__(self, first_name, last_name, username, email, age, location):
        self.first_name = first_name
        self.last_name = last_name
        self.username = username
        self.email = email
        self.age = age
        self.location = location
        self.__login_attempt = 0

    def describe_user(self):
        print("Informasi Pengguna:")
        print(f" Nama             : {self.first_name} {self.last_name}")
        print(f" Username         : {self.username}")
        print(f" Email            : {self.email}")
        print(f" Umur             : {self.age}")
        print(f" Lokasi           : {self.location}")
        print(f" Login Attempt    : {self.__login_attempt}")
        print(f" Max Login Attempt: {User.__max_login_attempt}")

    def greet_user(self):
        print(f"Halo {self.first_name}, selamat datang kembali!")

    def increment_login_attempt(self):
        self.__login_attempt += 1
        if self.__login_attempt > User.__max_login_attempt:
            print(f"Notifikasi: Anda telah melebihi batas maksimal login attempt ({User.__max_login_attempt}).")
        else:
            print(f"Login attempt sekarang: {self.__login_attempt}")

    def reset_login_attempt(self):
        self.__login_attempt = 0
        print("Login attempt telah di-reset ke 0.")

    @property
    def login_attempt(self):
        return self.__login_attempt

    @classmethod
    def set_max_login_attempt(cls, new_max):
        cls.__max_login_attempt = new_max
        print(f"Nilai max login attempt telah diubah menjadi {new_max}.")


restaurant1 = Restaurant("Layar Seafood", "Surabaya")
restaurant2 = Restaurant("Scarlette", "Jakarta")

restaurant1.add_menu("Cumi Bakar", 50000)
restaurant1.add_menu("Udang Sambal Matah", 60000)
restaurant1.add_menu("Ikan Bakar", 55000)

restaurant2.add_menu("Pasta Carbonara", 85000)
restaurant2.add_menu("Pizza Margherita", 75000)
restaurant2.add_menu("Risotto", 90000)

restaurant1.describe_restaurant()
restaurant1.display_menu()
print()
restaurant2.describe_restaurant()
restaurant2.display_menu()
restaurant1.open_restaurant()
restaurant2.open_restaurant()

print("\n------------------------------------\n")

user1 = User("Cio", "Siregar", "cio_s", "cio@happy.com", 25, "Jakarta")
user2 = User("Cia", "Andini", "cia_a", "cia@happy.com", 23, "Surabaya")

print("Pengguna 1:")
user1.describe_user()
user1.greet_user()

print("\nPengguna 2:")
user2.describe_user()
user2.greet_user()

print("\n--- Uji Coba Login Attempt ---")
print(f"Login attempt awal untuk {user1.username}: {user1.login_attempt}")

user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()
user1.increment_login_attempt()

print(f"Total login attempt untuk {user1.username} setelah increment: {user1.login_attempt}")

user1.reset_login_attempt()
print(f"Login attempt untuk {user1.username} setelah reset: {user1.login_attempt}")

User.set_max_login_attempt(5)
print("\nSetelah mengubah max login attempt:")
user1.describe_user()
user2.describe_user()
```
