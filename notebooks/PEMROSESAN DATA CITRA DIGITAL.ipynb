```python
#2. EKSTRAKSI BAND
b, g, r = cv2.split(img)
b = img[...,0]
g = img[...,1]
r = img[...,2]
cv2_imshow(b);
```

```python
#3 TRANFORMASI COLORSPACE
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV);
h,s,v = cv2.split(hsv)
cv2_imshow(h)
```

```python
#4.1 MENAMPILKAN HISTOGRAM CITRA
import cv2
import numpy as np
from matplotlib import pyplot as plt
color = ('b', 'g', 'r')
for i, col in enumerate(color):
  histr = cv2.calcHist([img], [i], None, [256], [0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
plt.show()
```

```python
#5.1 CONTOH KASUS: EKSTRAKSI OBJEK DENGAN WARNA MENGGUNAKAN OpenCV
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('download.jpeg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV);
batas_bawah = 46
batas_atas = 78
lower_orange = np.array([batas_bawah, 50, 50])
upper_orange = np.array([batas_atas, 255, 255])
mask = cv2.inRange(hsv, lower_orange, upper_orange)
res = cv2.bitwise_and(img, img, mask=mask)
print("gambar asal")
cv2_imshow(img);
print("masking warna")
cv2_imshow(mask)
print("hasil operasi")
cv2_imshow(res)
```
