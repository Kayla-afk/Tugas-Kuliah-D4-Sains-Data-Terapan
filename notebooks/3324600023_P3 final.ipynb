```python
#latihan 2 
def menu():
    print ("\nMenu:")
    print("1. Tulis ke File")
    print("2. Baca dari File")
    print("3. Keluar")
    return input("Pilih opsi (1/2/3): ")

def tulis_ke_file():
    nama_file = "e:/SEMESTER 2/PEMROGRAMAN 2 (BU LELY)/coba 6 maret.txt"
    with open(nama_file, 'w') as file: 
        teks = input("Masukkan teks: ")
        file.write(teks)  
        print("Teks berhasil ditulis ke file.")

def baca_dari_file():
    nama_file = "e:/SEMESTER 2/PEMROGRAMAN 2 (BU LELY)/coba 6 maret.txt"
    try:
        with open(nama_file, 'r') as file:  
            print("Isi file:")
            print(file.read())  
    except FileNotFoundError:
        print("File tidak ditemukan. Harap buat file terlebih dahulu.")

while True:
    pilihan = menu()
    if pilihan == '1':
        tulis_ke_file()
    elif pilihan == '2':
        baca_dari_file()
    elif pilihan == '3':
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak valid.")
```

```python
#latihan 3
def simpan_nama(nama_baru, nama_file):
    try:
        with open(nama_file, "r") as file:
            daftar_nama = file.read().splitlines()
    except FileNotFoundError:
        daftar_nama = []

    if nama_baru in daftar_nama:
        print(f"Nama '{nama_baru}' sudah ada di daftar.")
    else:
        with open(nama_file, "a") as file:
            file.write(nama_baru + "\n")
        print(f"Nama '{nama_baru}' berhasil disimpan.")

nama_file = "daftar_nama.txt"
nama_baru = input("Masukkan nama yang ingin disimpan: ")
simpan_nama(nama_baru, nama_file)
```

```python
#latihan 4
import os

def buat_file_baru(nama_file):
    files_di_direktori = os.listdir()

    if nama_file in files_di_direktori:
        print(f"File '{nama_file}' sudah ada di direktori.")
    else:
        with open(nama_file, "w") as file:
            file.write("")  
        print(f"File '{nama_file}' berhasil dibuat.")

nama_file_baru = input("Masukkan nama file baru (dengan ekstensi .txt): ")
buat_file_baru(nama_file_baru)
```
