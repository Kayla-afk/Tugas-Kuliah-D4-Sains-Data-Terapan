```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

# Dataset dari tabel
data = {
    "Panjang": ["Panjang","Panjang","Panjang",
                "Sedang","Sedang","Sedang",
                "Kecil","Kecil","Kecil"],
    "Lebar": ["kecil","sedang","Lebar",
              "kecil","sedang","Lebar",
              "kecil","sedang","Lebar"],
    "Berat": ["ringan","Berat","Berat",
              "Berat","ringan","Berat",
              "ringan","ringan","Berat"],
    "Jenis": ["Ikan Lele","Ikan Lele","Gurami",
              "Ikan Lele","Ikan Lele","Gurami",
              "Gurami","Gurami","Gurami"]
}

# Masukkan ke dataframe
df = pd.DataFrame(data)

# Encode data kategorikal ke numerik
le_Panjang = LabelEncoder()
le_Lebar = LabelEncoder()
le_Berat = LabelEncoder()
le_Jenis = LabelEncoder()

df["Panjang_enc"] = le_Panjang.fit_transform(df["Panjang"])
df["Lebar_enc"]   = le_Lebar.fit_transform(df["Lebar"])
df["Berat_enc"]   = le_Berat.fit_transform(df["Berat"])
df["Jenis_enc"]   = le_Jenis.fit_transform(df["Jenis"])

# Fitur dan label
X = df[["Panjang_enc","Lebar_enc","Berat_enc"]]
y = df["Jenis_enc"]

# Model Naive Bayes
model = CategoricalNB()
model.fit(X, y)

# Data yang akan diprediksi: Panjang, Lebar, Berat
sample = pd.DataFrame({
    "Panjang_enc": le_Panjang.transform(["Panjang"]),
    "Lebar_enc":   le_Lebar.transform(["Lebar"]),
    "Berat_enc":   le_Berat.transform(["Berat"])
})

# Prediksi
pred = model.predict(sample)[0]
jenis_prediksi = le_Jenis.inverse_transform([pred])[0]

print("Prediksi ikan:", jenis_prediksi)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

# Dataset dari tabel
data = {
    "Umur": ["<=30","<=30","30...40",">40",">40",">40","31...40","<=30","<=30",">40","<=30","31...40","31...40",">40"],
    "Gaji": ["tinggi","tinggi","tinggi","sedang","rendah","rendah","rendah","sedang","rendah","sedang","sedang","sedang","tinggi","sedang"],
    "Status": ["single","single","single","single","menikah","menikah","menikah","single","menikah","menikah","menikah","single","menikah","single"],
    "Hutang": ["punya","tidak","punya","punya","punya","tidak","tidak","punya","punya","punya","tidak","tidak","punya","tidak"],
    "BeliLaptop": ["tidak","tidak","iya","iya","iya","tidak","iya","tidak","iya","iya","iya","iya","iya","tidak"]
}

# Masukkan ke dataframe
df = pd.DataFrame(data)

# Encode data kategorikal ke numerik
le_Umur = LabelEncoder()
le_Gaji = LabelEncoder()
le_Status = LabelEncoder()
le_Hutang = LabelEncoder()
le_BeliLaptop = LabelEncoder()

df["Umur_enc"] = le_Umur.fit_transform(df["Umur"])
df["Gaji_enc"] = le_Gaji.fit_transform(df["Gaji"])
df["Status_enc"] = le_Status.fit_transform(df["Status"])
df["Hutang_enc"] = le_Hutang.fit_transform(df["Hutang"])
df["BeliLaptop_enc"] = le_BeliLaptop.fit_transform(df["BeliLaptop"])

# Fitur dan label
X = df[["Umur_enc","Gaji_enc","Status_enc","Hutang_enc"]]
y = df["BeliLaptop_enc"]

# Model Naive Bayes
model = CategoricalNB()
model.fit(X, y)

# Data yang akan diprediksi
sample = pd.DataFrame({
    "Umur_enc": le_Umur.transform(["<=30"]),
    "Gaji_enc": le_Gaji.transform(["sedang"]),
    "Status_enc": le_Status.transform(["menikah"]),
    "Hutang_enc": le_Hutang.transform(["punya"])
})

# Prediksi
pred = model.predict(sample)[0]
pred_label = le_BeliLaptop.inverse_transform([pred])[0]

# Probabilitas prediksi
proba = model.predict_proba(sample)[0]

print("Prediksi Beli Laptop:", pred_label)
print("Probabilitas [Tidak, Iya]:", proba)
```
