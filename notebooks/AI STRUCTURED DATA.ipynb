```python
# LAB 1: Structured vs Unstructured Data

import pandas as pd
data_terstruktur = {
    "ID": [1, 2, 3],
    "Nama": ["Kayla", "Milo", "Wooly"],
    "Umur": [20, 5, 6]
}
df = pd.DataFrame(data_terstruktur)
print("=== Structured Data (DataFrame) ===")
print(df)

data_tidak_terstruktur = [
    "Kayla suka memasak.",
    "Milo senang minum susu.",
    "Wooly gemar mengeong."
]
print("\n=== Unstructured Data (Teks) ===")
for i, kalimat in enumerate(data_tidak_terstruktur, start=1):
    print(f"Dokumen {i}: {kalimat}")
```

```python
# LAB 2: Bag of Words (BoW)

from sklearn.feature_extraction.text import CountVectorizer

dokumen = [
    "Alice suka membaca buku dan menulis cerita.",
    "Bob senang bermain sepak bola di akhir pekan.",
    "Charlie gemar memasak dan mencoba resep baru."
]
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(dokumen)

vocab = vectorizer.get_feature_names_out()
print("Vocabulary:", vocab)
print("\nRepresentasi BoW (Matriks):")
print(X_bow.toarray())
```

```python
# LAB 3: TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(dokumen)

vocab_tfidf = tfidf_vectorizer.get_feature_names_out()
print("Vocabulary TF-IDF:", vocab_tfidf)
print("\nRepresentasi TF-IDF (Matriks):")
print(X_tfidf.toarray())
```
