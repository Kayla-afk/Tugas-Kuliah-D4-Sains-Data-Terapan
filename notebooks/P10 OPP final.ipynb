```python
#Exercise 1
class Car:
    vehicle_type = "Land"

    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def start_engine(self):
        return f"The engine of {self.brand} {self.model} is started!"

    def get_info(self):
        return f"Brand: {self.brand}, Model: {self.model}, Year: {self.year}"


class Airplane:
    vehicle_type = "Air"

    def __init__(self, airline, capacity, wingspan):
        self.airline = airline
        self.capacity = capacity
        self.wingspan = wingspan
    def take_off(self):
        return f"The airplane operated by {self.airline} is taking off!"
    def get_info(self):
        return f"Airline: {self.airline}, Capacity: {self.capacity}, Wingspan: {self.wingspan} meters"

car1 = Car("Toyota", "Corolla", 2022)
print(car1.start_engine())
print(car1.get_info())

plane1 = Airplane("Garuda Indonesia", 300, 60)
print(plane1.take_off())
print(plane1.get_info())

'''
Penjelasan:
1. Penjelasan Kelas Car
Kelas ini merepresentasikan objek mobil. Atribut dan metode yang dimiliki adalah:
Atribut Kelas: vehicle_type, yang berlaku untuk semua objek mobil. Nilainya "Land".
Atribut Instance: brand, model, dan year, merepresentasikan merek, model, dan tahun produksi mobil.
  Method:
    start_engine(): Method untuk menyalakan mesin mobil.
    get_info(): Method untuk mengambil informasi detail tentang mobil.

2. Penjelasan Kelas Airplane
Kelas ini merepresentasikan objek pesawat terbang. Atribut dan metode yang dimiliki adalah:
Atribut Kelas: vehicle_type, yang berlaku untuk semua objek pesawat. Nilainya "Air".
Atribut Instance: airline, capacity, dan wingspan, merepresentasikan maskapai, kapasitas penumpang, dan lebar sayap pesawat.
  Method:
    take_off(): Method untuk mensimulasikan pesawat lepas landas.
    get_info(): Method untuk mengambil informasi detail tentang pesawat.
```

```python
#Exercise 2
class Car:
    vehicle_type = "Land"

    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def start_engine(self):
        return f"The engine of {self.brand} {self.model} is started!"

    def get_info(self):
        return f"Brand: {self.brand}, Model: {self.model}, Year: {self.year}"


class Airplane:
    vehicle_type = "Air"

    def __init__(self, airline, capacity, wingspan):
        self.airline = airline
        self.capacity = capacity
        self.wingspan = wingspan

    def take_off(self):
        return f"The airplane operated by {self.airline} is taking off!"

    def get_info(self):
        return f"Airline: {self.airline}, Capacity: {self.capacity}, Wingspan: {self.wingspan} meters"


car1 = Car("Toyota", "Corolla", 2022)
car2 = Car("Honda", "Civic", 2021)

plane1 = Airplane("Garuda Indonesia", 300, 60)
plane2 = Airplane("Lion Air", 180, 55)

print(car1.get_info())
print(f"Vehicle Type: {Car.vehicle_type}")
print(car1.start_engine())

print(car2.get_info())
print(car2.start_engine())

print(plane1.get_info())
print(f"Vehicle Type: {Airplane.vehicle_type}")
print(plane1.take_off())

print(plane2.get_info())
print(plane2.take_off())

#Penjelasan:
#Dalam Python, kita bisa menggunakan konsep Object-Oriented Programming (OOP) untuk membuat program lebih terstruktur dan modular.
#Contohnya, kita mendefinisikan dua kelas, yaitu Car (mobil) dan Airplane (pesawat) dengan atribut-atribut dan metode masing-masing.

#1. Penjelasan Kelas Car
#Kelas ini mewakili objek mobil dengan atribut:
#Atribut Kelas: vehicle_type yang bernilai "Land". Atribut ini berlaku untuk semua objek dari kelas Car.
#Atribut Instance: brand, model, dan year, yaitu informasi spesifik tentang merek, model, dan tahun produksi mobil.
#Method: start_engine() untuk menyalakan mesin dan get_info() untuk mendapatkan informasi detail tentang mobil.

#2. Penjelasan Kelas Airplane
#Kelas ini mewakili objek pesawat dengan atribut:
#Atribut Kelas: vehicle_type yang bernilai "Air". Atribut ini berlaku untuk semua objek dari kelas Airplane.
#Atribut Instance: airline, capacity, dan wingspan, yaitu informasi spesifik tentang maskapai, kapasitas penumpang, dan lebar sayap pesawat.
#Method: take_off() untuk melakukan simulasi pesawat lepas landas dan get_info() untuk mendapatkan informasi tentang pesawat.
```
