```python
#1. Membaca Dataset
import pandas as pd
import numpy as np

file_path = '/content/BANK LOAN.csv'
data = pd.read_csv(file_path)

print("Preview Dataset:")
print(data.head())

print("\nInformasi Dataset:")
print(data.info())

print("\nStatistik Deskriptif:")
print(data.describe())

print("\nCek Nilai Hilang:")
print(data.isnull().sum())
```

```python
#2 Preprocessing data
#Mengatasi missing value
from sklearn.impute import SimpleImputer
import pandas as pd

data = {'column': [1, 2, None, 4, 5]}
df = pd.DataFrame(data)

imputer = SimpleImputer(strategy='mean')

df['column'] = imputer.fit_transform(df[['column']])

#Encoding kategori
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data = {'category_column': ['A', 'B', 'A', 'C']}
df = pd.DataFrame(data)
encoder = OneHotEncoder(sparse_output=False)
transformed = encoder.fit_transform(df[['category_column']])
encoded_df = pd.DataFrame(
    transformed,
    columns=encoder.get_feature_names_out(['category_column'])
)
print(encoded_df)

```

```python
#3. Normalisasi data
#Min-max xcalling
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
data = {'column': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
scaler = MinMaxScaler()
df['normalized_column'] = scaler.fit_transform(df[['column']])
print(df)

#Z-score normalization
from sklearn.preprocessing import StandardScaler
import pandas as pd
data = {'column': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
scaler = StandardScaler()
df['zscore_normalized'] = scaler.fit_transform(df[['column']])

print(df)

```

```python
#4. Mengatasi outlier
#Identifikasi outlier dengan IQR
import pandas as pd
data = {'column': [10, 12, 15, 18, 20, 200]}
df = pd.DataFrame(data)
Q1 = df['column'].quantile(0.25)
Q3 = df['column'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['column'] < lower_bound) | (df['column'] > upper_bound)]
print("Outlier:")
print(outliers)

#Mengatasi outlier dengan capping
df['column_capped'] = df['column'].clip(lower=lower_bound, upper=upper_bound)
print("\nData setelah Capping:")
print(df)

#Transformasi data
import numpy as np
df['column_log'] = np.log1p(df['column'])
print("\nData setelah Transformasi Log:")
print(df)

```
