# **Split Dataset menjadi Atribut dan Target**

```python
target = data['Berolah-raga'].values
atribut = data.drop('Berolah-raga', axis=1).values
```

# **Inisialisasi Hipotesis Spesifik**

```python
H = ['0']*atribut.shape[1]
```

```python
H
```

# **Implementasi Logika dari Algoritma Find-S**

```python
for i in range(atribut.shape[0]):
  if target[i] == 'Ya':
    for j in range(atribut.shape[1]):
      if H[j] == '0':
        H[j] = atribut[i][j]
      elif H[j] != atribut[i][j]:
        H[j] = '?'
    print(f'For Step - {i} : {H}')
```

# **Final Hipotesis Awal**

```python
print(f'Final Hipotesis : {H}')
```
