Buatlah class Restaurant dengan dua atribut, yakni name dan cuisine_type.

# **LATIHAN 2**
Buatlah class IceCreamStall yang merupakan anak dari class  Restaurant.
Lengkapi class tersebut dengan atribut flavors yang berisi daftar rasa es krim yang tersedia.
Lengkapi pula class tersebut dengan method show_flavors() yang bertugas untuk menampilkan ke layar daftar rasa yang tersedia.
Buat obyek dari class IceCreamStall dan Restaurant. Lakukan uji coba untuk mengakses semua method dan atribut yang ada pada kedua class tersebut.


```python
class Restaurant:
    def __init__(self, name, cuisine_type):
        self.name = name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        print(f"Restoran '{self.name}' menyajikan hidangan {self.cuisine_type}.")

    def open_restaurant(self):
        print(f"Restoran '{self.name}' sekarang dibuka!")

# Membuat subclass IceCreamStall
class IceCreamStall(Restaurant):
    def __init__(self, name, cuisine_type, flavors):
        super().__init__(name, cuisine_type)
        self.flavors = flavors

    def show_flavors(self):
        print(f"Rasa es krim yang tersedia di '{self.name}': {', '.join(self.flavors)}")

# Membuat objek dari class Restaurant
restoran1 = Restaurant("Remboelan", "masakan Indonesia")
restoran2 = Restaurant("Gyukaku", "masakan Jepang")

# Membuat objek dari class IceCreamStall
ice_cream_stall = IceCreamStall("Sweet Treats", "es krim", ["matcha", "greekyougurt", "avocoffee"])

# Mengakses metode describe_restaurant untuk semua objek
restoran1.describe_restaurant()
restoran2.describe_restaurant()
ice_cream_stall.describe_restaurant()

# Membuka restoran
ice_cream_stall.open_restaurant()

# Menampilkan daftar rasa es krim yang tersedia
ice_cream_stall.show_flavors()
```

# **LATIHAN 5**
Buatlah class Admin yang menuruni sifat dari class User.
Tambahkan atribut privileges pada class tersebut yang berisi daftar privilege, antara lain: “dapat menambahkan post”, “dapat menghapus post”, “dapat melarang user”.
Tambahkan method show_privileges() untuk menampilkan daftar privileges yang dimiliki oleh admin.
Buat obyek dari class User dan Admin. Lakukan uji coba untuk mengakses semua method dan atribut yang ada pada kedua class tersebut.


```python
class User:
    def __init__(self, first_name, last_name):
        self.__first_name = first_name
        self.__last_name = last_name
    def describe_user(self):
        print(f"User: {self.__first_name} {self.__last_name}")

    def greet_user(self):
        print(f"Morning {self.__first_name}, Welcome dear!")

class Admin(User):
    def __init__(self, first_name, last_name):
        super().__init__(first_name, last_name)
        self.privileges = [
            "dapat menambahkan post",
            "dapat menghapus post",
            "dapat melarang user"
        ]

    def show_privileges(self):
        print(f"Admin '{self._User__first_name}' memiliki hak akses berikut:")
        for privilege in self.privileges:
            print(f"- {privilege}")

user1 = User("Cia", "Sheeds")
user2 = Admin("Juyeon", "isCat")

user1.describe_user()
user1.greet_user()

user2.describe_user()
user2.greet_user()

user2.show_privileges()
```

# **LATIHAN 6**
Modifikasi method describe_user() pada class Admin yang berfungsi untuk mencetak informasi terkait user, termasuk daftar privilege yang dimiliki oleh objek dari class Admin .


```python
class User:
    def __init__(self, first_name, last_name):
        self.__first_name = first_name
        self.__last_name = last_name

    def describe_user(self):
        print(f"User: {self.__first_name} {self.__last_name}")

    def greet_user(self):
        print(f"Morning {self.__first_name}, Welcome dear!")

class Admin(User):
    def __init__(self, first_name, last_name):
        super().__init__(first_name, last_name)
        self.privileges = [
            "dapat menambahkan post",
            "dapat menghapus post",
            "dapat melarang user"
        ]

    def describe_user(self):
        print(f"Admin: {self._User__first_name} {self._User__last_name}")
        print("Hak akses yang dimiliki:")
        for privilege in self.privileges:
            print(f"- {privilege}")

    def show_privileges(self):
        print(f"Admin '{self._User__first_name}' memiliki hak akses berikut:")
        for privilege in self.privileges:
            print(f"- {privilege}")

user1 = User("Cia", "Sheeds")
user2 = Admin("Juyeon", "isCat")

user1.describe_user()
user1.greet_user()

user2.describe_user()
user2.greet_user()
```

# **LATIHAN 7**
Buatlah program sederhana yang mengimplementasikan konsep inheritance dan polymorphism (overriding).


```python
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("Suara hewan")

class Dog(Animal):
    def make_sound(self):
        print(f"{self.name} menggonggong: Woof! Woof!")

class Cat(Animal):
    def make_sound(self):
        print(f"{self.name} mengeong: Meow! Meow!")

anjing = Dog("Bridney")
kucing = Cat("Lulu")

anjing.make_sound()
kucing.make_sound()
```
