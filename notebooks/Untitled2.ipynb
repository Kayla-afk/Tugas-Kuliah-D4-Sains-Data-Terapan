```python
import skfuzzy.control as ctrl
import numpy as np

service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
food = ctrl.Antecedent(np.arange(0, 11, 1), 'food')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')

import skfuzzy as fuzz
service['poor'] = fuzz.trimf(service.universe, [0, 0, 5])
service['average'] = fuzz.trimf(service.universe, [0, 5, 10])
service['good'] = fuzz.trimf(service.universe, [5, 10, 10])

import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl

food = ctrl.Antecedent(np.arange(0, 11, 1), 'food')

food['poor'] = fuzz.trimf(food.universe, [0, 0, 5])
food['average'] = fuzz.trimf(food.universe, [0, 5, 10])
food['good'] = fuzz.trimf(food.universe, [5, 10, 10])

##
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl

# Define input variables
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
food = ctrl.Antecedent(np.arange(0, 11, 1), 'food')

# Define output variable
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')

# Define membership functions for all variables
service['poor'] = fuzz.trimf(service.universe, [0, 0, 5])
service['average'] = fuzz.trimf(service.universe, [0, 5, 10])
service['good'] = fuzz.trimf(service.universe, [5, 10, 10])

food['poor'] = fuzz.trimf(food.universe, [0, 0, 5])
food['average'] = fuzz.trimf(food.universe, [0, 5, 10])
food['good'] = fuzz.trimf(food.universe, [5, 10, 10])

tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])

# Define fuzzy rules
rule1 = ctrl.Rule(service['poor'] | food['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] & food['good'], tip['high'])

# Ensure rules are defined before using them
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)

# Now you can use these rules in your simulation
tipping.input['service'] = 8
tipping.input['food'] = 6
tipping.compute()

print(tipping.output['tip'])

tipping.input['service'] = 8.5
tipping.input['food'] = 7.0
tipping.compute()
print(tipping.output['tip'])

service.view()
tip.view(sim=tipping)
```
