```python
#EXERCISE 1
# test case:
# valid: anaaa, arumi, sania, vikar, andragogi
# invalid: anda, an_a, Arumi, sania99, diana
import re
texts = ["anaa", "arumi", "sania", "vikar", "andragogi",
         "anda", "an_a", "Arumi", "sania99", "diana"]
for text in texts:
    match = re.search("^[a-z]{,5}$", text)
    if match:
        print(f"'{text}' is valid")
    else:
        print(f"'{text}' is invalid")
```

```python
#EXERCISE 2
#valid: 11anaaa, 18arum, 30sania, 09vika, 43Andreas, 67dian.a
#invalid: andara11, _andri, Arumi10, 9sania, dian.a, 10Aruumi10
import re

usernames = ["11anaaa", "18arum", "30sania", "09vika", "43Andreas", "67dian.a", "andara11", "_andri", "Arumi10", "9sania", "dian.a", "10Aruumi10"]
pattern = r"^\d{2}[a-zA-Z.]+$"

for username in usernames:
    if re.fullmatch(pattern, username):
        print(f"'{username}' is valid")
    else:
        print(f"'{username}' is invalid")
```

```python
#EXERCISE 3
import re

emails = ["Deby@pens.ac.id", "amai78@pens.ac.id", "lely_aja@pens.ac.id", "d4@pens.ac.id", "7dian@pens.ac.id", "d@pens.ac.id", "deby@pens.com", "_deby@pens"]

pattern = r"^[a-zA-Z][a-zA-Z0-9_]{1,}@pens\.ac\.id$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(f"'{email}' is valid")
    else:
        print(f"'{email}' is invalid")
```
