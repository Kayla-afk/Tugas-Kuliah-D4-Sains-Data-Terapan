```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 1. Load data
file_path = "/IPM Jatim.xlsx"
data = pd.read_excel(file_path)

# 2. Pilih kolom yang relevan untuk clustering
features = data[['Indeks Kesehatan', 'Indeks Pendidikan', 'Indeks Ekonomi']]

# 3. Normalisasi data agar skala sebanding
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 4. Tentukan jumlah cluster (contoh: 3 cluster)
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# 5. Tampilkan hasil cluster per kota
print(data[['City', 'Cluster']])

# 6. Visualisasi cluster (2D, menggunakan 2 indeks)
plt.figure(figsize=(8,6))
plt.scatter(data['Indeks Kesehatan'], data['Indeks Pendidikan'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Indeks Kesehatan')
plt.ylabel('Indeks Pendidikan')
plt.title('Clustering Kabupaten/Kota Jawa Timur berdasarkan IPM')
plt.colorbar(label='Cluster')
plt.show()
```
