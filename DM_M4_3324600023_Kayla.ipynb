```python
import pandas as pd #untuk memanipulasi dataset
import numpy as np #untuk membantu dalam perhitungan numerik
```

```python
#1. Menampilkan dataset
dataset = pd.read_csv('/content/titanic.csv')
print("Dataset Titanic: ")
dataset
```

```python
print(dataset.columns)
```

```python
#2. Menampilkan jumlah baris dan kolom
rows, cols = dataset.shape #shape digunakan untuk mengembalikan (rows, column)
print("Jumlah baris: ", rows)
print("Jumlah kolom: ", cols)
```

```python
#3. Megambil fitur (Age, Fare)
dataset = dataset[['Age', 'Fare']]
print("Fitur (Age dan Fare):")
dataset
```

```python
#4. Mengambil kelas Survived
import pandas as pd

original_dataset = pd.read_csv('/content/titanic.csv')
kelas = original_dataset['Survived']

print("Class (Survived):")
kelas
```

```python
#5. Mengisi missing value age berdasarkan mean masing-masing class
dataset['Survived'] = kelas

dataset['Age'] = dataset.groupby('Survived')['Age'].transform(
    lambda x: x.fillna(x.mean())
) #nilai age yang kosong diisi dengan mean dari class tersebut

print("Dataset setelah pengisian missing value Age:")
dataset[['Age', 'Fare', 'Survived']]
```

```python
# Ambil data fitur setelah data preprocessing
data = dataset[['Age', 'Fare']]

print("Data setelah preprocessing:")
data
```

```python
#6. Normalisasi min-max
min_val = data.min()
max_val = data.max()

data_minmax = (data - min_val) / (max_val - min_val)

print("Hasil Normalisasi Min-Max (0-1):")
data_minmax
```

```python
#7. Normalisasi Z-Score
mean = data.mean()
std = data.std()

data_zscore = (data - mean) / std

print("Hasil Normalisasi Z-score:")
data_zscore
```

```python
print("Mean setelah Z-score:")
print(data_zscore.mean())

print("\nStd setelah Z-score:")
print(data_zscore.std())
```

```python
#8. Normalisasi Sigmoidal
data_sigmoid = 1 / (1 + np.exp(-data))
print("Hasil Normalisasi Sigmoidal:")
data_sigmoid
```
